```python
import requests
import json
from pprint import pprint

# get secret from Colab
from google.colab import userdata
token = userdata.get('git_token')

# initialize request parameters
url = 'https://api.github.com/user'
headers = {'Authorization': 'Bearer ' + token}

# send request to GitHub
r = requests.get(url, headers=headers)

# print initial responses
print(r.status_code)
print(r.text)

r_json = json.loads(r.text)
pprint(r_json)
```

```python
# initialize request parameters
url = 'https://api.github.com/search/repositories?q=stars&sort=stars&per_page=10'
headers = {'Authorization': 'Bearer ' + token}

# send request to GitHub
r = requests.get(url, headers=headers)

# print initial responses
print(r.status_code)
print(r.text)

r_json = json.loads(r.text)
pprint(r_json)
```

```python
items = r_json['items']
print(items[0])

repo_name = []
repo_stars = []
for item in items:
  repo_name.append(item['name'])
  repo_stars.append(item['stargazers_count'])

print(repo_name)
print(repo_stars)
```

```python
import matplotlib.pyplot as plt

plt.barh(repo_name, repo_stars)
plt.title('Top 10 repo users')
plt.show();
```

```python
# initialize request parameters
url = 'https://api.github.com/search/users?q=followers&sort=followers&per_page=10'
headers = {'Authorization': 'Bearer ' + token}

# send request to GitHub
r = requests.get(url, headers=headers)

# print initial responses
print(r.status_code)
print(r.text)

r_json = json.loads(r.text)
pprint(r_json)
```

```python
items = r_json['items']
print(items[0])

user_name = []
user_score = []
for item in items:
  user_name.append(item['login'])
  user_score.append(item['score'])

print(user_name)
print(user_score)
```

```python
plt.barh(user_name, user_score)
plt.title('Score of top 10 users')
plt.show();
```
